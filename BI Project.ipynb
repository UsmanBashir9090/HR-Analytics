# Specialization Column

Making relationship of Highest Qualification and Specialization

```python
table3 = pd.pivot_table(df,index='highest qualification', columns='specialization', values='candidateID',aggfunc='count', fill_value=0,
                         margins=True,
                         margins_name="Total")
table3
```

As pivot table shows we have 2479 missing values in total 

```python
func = lambda x: round(100*x.count()/df.shape[0],1)
pd.pivot_table(df, index='highest qualification',columns='specialization' , values='candidateID', aggfunc=func,
                         margins=True,
                         margins_name="Total")
```

As from the pivot table we can analyze that High School and Primary School has no specialization so we can categorized them in No major 

Which leaves us with only 1.1% missing values which can be replaced with STEM as it dominates the column with 76.9 % including missing values. 

```python
for index in df.index:
 #if graduate,phd or masters then stem  
    if (df.loc[index, 'highest qualification']  == 'Graduate' or df.loc[index, 'highest qualification']  == 'Phd' or df.loc[index, 'highest qualification']  == 'Masters') and  df.loc[index, 'specialization'] == 'mv' :
        df.loc[index, 'specialization'] = 'STEM'
 # if high school or primary then no major       
    elif (df.loc[index, 'highest qualification']  == 'High School' or df.loc[index, 'highest qualification']  == 'Primary School') and  df.loc[index, 'specialization'] == 'mv':
        df.loc[index, 'specialization'] = 'No Major'
         
```

Missing Values Replaced

```python
#Fixing hidden repetition of data
explore_values = df[['specialization']]
#Step 1 -Recall the replacements needed to be made
for col in explore_values:
    print(explore_values[col].unique())
df['specialization'].value_counts()
```
